## 控制AI行为

_这是2025年日本神奈川县的社会保险金额表格，请帮我分析这个表格中各项保险的算法，要考虑表格及其说明文字提及的各种特殊情况、计算的边界，不仅要告诉我各种保险的计算公式，还有会影响到计算结果的各种因素。请注意，仅分析，在我确认之前，不要开始编写代码。_

## 描述技术栈

请按以下要求进行重构：

① 使用React构建主要的前端框架。  
② 使用Redux管理状态。  
③ 使用Next.js以提供更好的用户体验和性能。  
④ 使用Tailwind CSS实现视觉效果。  
⑤ 使用微软的Fluent交互风格（或谷歌的Material交互风格）。  

## 发布

① 将代码托管到Github。  
② 将原型产品发布到Cloud Run。  
③ 预览原型。  

## 持续交付(CD)

① 在Google Cloud生态中实现持续交付。  
② 通过Github + Google Cloud Run实现持续交付。

## 作业

① 按照本节所学的知识开发一个计算社会保险金的原型产品。  
② 向原型产品添加雇用保险数据。  
③ 向原型产品添加源泉征收税数据。
